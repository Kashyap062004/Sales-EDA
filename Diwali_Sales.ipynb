```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# deleting colomns "Status" and "Unnamed1"
df=df.drop(['Status','unnamed1'],axis=1)
```

```python
df.info()
```

```python
# no of null values in colomn
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# Remove all null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python

# make amount whole number
df['Amount']=df['Amount'].astype('int')
```

```python
df
```

```python

df.columns
```

```python
# rename of column Cust_name to Customer_Name
df=df.rename(columns={'Cust_name':'Customer_Name'})
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
# Genderwise sales
Genderwise_sales=sns.countplot(x='Gender',data=df)
for i in Genderwise_sales.containers:
    Genderwise_sales.bar_label(i)
```

```python
# Age-wise purchase by unit
sns.countplot(data=df,x='Age Group')
```

```python
# Age-wise and Gender-wise purchase by unit
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
# Age-wise  purchase by amount
Sales_agewise_by_amount=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Sales_agewise_by_amount)
```

From the above we can conclude that costumer of age between 26-35 as bought items of maximum amount 

```python
# State wise 
state_wise_sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=state_wise_sales,x='State',y='Amount')
```

```python
# State wise  Amount of top 5 States
state_wise_sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(data=state_wise_sales,x='State',y='Amount')
```

```python
# State wise  order of top 5 States
state_wise_sales=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head()
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(data=state_wise_sales,x='State',y='Orders')
```

```python
# Marrage status
Marrage_status=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for i in Marrage_status.containers:
    Marrage_status.bar_label(i)
```

```python
# Marrage status and genderwise
Marrage_status_and_genderwise=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=Marrage_status_and_genderwise,x='Marital_Status',y='Amount',hue='Gender')
```

the maximum buyer are married women also they have higer purchesing power

```python
# Occupation 
sns.set(rc={'figure.figsize':(20,5)})
Occupationwise=sns.countplot(data=df,x='Occupation')
for i in Occupationwise.containers:
    Occupationwise.bar_label(i)
```

```python
occupationwise=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


sns.barplot(data=occupationwise,x='Occupation',y='Amount')
```
